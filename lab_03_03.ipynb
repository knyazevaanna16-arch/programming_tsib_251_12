Исрпавленная программа

```python
# Задание task_03_03_01.
#
# Выполнил: Князева А. П.
# Группа: ЦИБ-251

"""
Ошибки (номера строк через пробел, данная строка - №2): 9 11 13 15
"""


def sum_of_digits(n):
    """Вернуть сумму цифр меньших 5 для положительного целого числа `n`.
    Если таких цифр нет, вернуть 0."""
    s = 0
    while n > 0:
        digit = n % 10
        if digit < 5:
            s = s + digit
        n //= 10
    return s
```
