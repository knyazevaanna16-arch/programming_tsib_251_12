Исрпавленная программа

```python
# Задание task_03_03_01.
#
# Выполнил: Князева А. П.
# Группа: ЦИБ-251

"""
Ошибки (номера строк через пробел, данная строка - №2): 7 9 11 13
"""
def sum_of_digits(n):
    """Вернуть сумму цифр меньших 5 для положительного целого числа `n`.
    Если таких цифр нет, вернуть 0.
    Исключения:
    - TypeError: если n не число
    - ValueError: если n не целое или отрицательное
    """
    if type(n) not in (int, float):
        raise TypeError("Аргумент должен быть числом")

    if n < 0:
        raise ValueError("Число должно быть положительным")

    if n != int(n):
        raise ValueError("Число должно быть целым")

    if n == 0:
        return 0

    total = 0
    n = int(n)

    while n > 0:
        digit = n % 10
        if digit < 5:
            total = total + digit
        n //= 10
    return total

# Демонстрация работы функции
print("Правильная работа:")
print(sum_of_digits(1234))
print(sum_of_digits(5678))
print(sum_of_digits(0))
print(sum_of_digits(5))
print(sum_of_digits(19))

print("\nИсключения:")

try:
    result = (sum_of_digits(-5))
except ValueError as e:
    print("Ошибка:", e)

try:
    result = (sum_of_digits(3.14))
except ValueError as e:
    print("Ошибка:", e)

try:
    result = (sum_of_digits("abc"))
except TypeError as e:
    print("Ошибка:", e)

try:
    result = (sum_of_digits(None))
except TypeError as e:
    print("Ошибка:", e)

try:
    result = (sum_of_digits([1, 2, 3]))
except TypeError as e:
    print("Ошибка:", e)
```
