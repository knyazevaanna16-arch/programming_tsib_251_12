```python
# task_04_01_12

# Домен / Сфера: Клиника; Базовый класс 1 (Сущность): MedicalService (стоим.);
# Базовый класс 2 (Субъект): Patient (карта)

# Выполнил: Князева А.П.
# Группа ЦИБ 251

class MedicalService:
    """Класс для описания медицинской услуги."""

    def __init__(self, name, cost):
        """
        Конструктор класса.
        :param name: Название услуги (строка)
        :param cost: Стоимость услуги (число)
        """
        self.name = name
        self.cost = cost

    def __str__(self):
        """Возвращает красивое строковое представление объекта."""
        return f"Услуга: {self.name:<17} | Стоимость: {self.cost} руб."


class Patient:
    """Класс для описания пациента."""

    def __init__(self, full_name, card_number):
        """
        Конструктор класса.
        :param full_name: ФИО пациента (строка)
        :param card_number: Номер медицинской карты (строка или число)
        """
        self.full_name = full_name
        self.card_number = card_number

    def __str__(self):
        """Возвращает красивое строковое представление объекта."""
        return f"Пациент: {self.full_name:<27} | Карта №: {self.card_number}"


if __name__ == "__main__":
    # --- Демонстрация работы классов ---

    # 1. Создаем объекты медицинских услуг (Базовый класс 1)
    service_1 = MedicalService("Прием терапевта", 2800)
    service_2 = MedicalService("Анализ крови", 2200)
    service_3 = MedicalService("УЗИ сердца", 2500)

    # 2. Создаем объекты пациентов (Базовый класс 2)
    patient_1 = Patient("Иванова Мария Петровна", "123-456")
    patient_2 = Patient("Смирнов Алексей Игоревич", "789-012")

    # 3. Выводим информацию об объектах в консоль
    # Благодаря методу __str__ вывод будет читаемым ("Vibe Code")
    print("--- Список услуг ---")
    print(service_1)
    print(service_2)
    print(service_3)

    print("\n--- Список пациентов ---")
    print(patient_1)
    print(patient_2)
```
