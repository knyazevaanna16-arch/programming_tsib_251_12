```python
# Задание task_02_12.
#
# Даны целые числа a и b (a может быть больше b). Напечатайте:
# числа от минимального до максимального в строчку (разделяя пробелом);
# числа от максимального до минимального «столбиком».
#
# Выполнил: Князева А. П.
# Группа: ЦИБ-251

a = int(input("a = "))
b = int(input("b = "))

# Определяем минимум и максимум
if a < b:
    minimum = a
    maximum = b
else:
    minimum = b
    maximum = a

# Числа от минимального до максимального в строку
for i in range(minimum, maximum + 1):
    print(i, end=" ")
print()  # переход на новую строку

# Числа от максимального до минимального столбиком
for i in range(maximum, minimum - 1, -1):
    print(i)
```
