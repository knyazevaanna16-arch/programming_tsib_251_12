```python
# Задание task_03_02_02.
#
# Выполнил: Князева А. П.
# Группа: ЦИБ-251

def foo(i):
    """ Преобразует целое неотрицательное число в строку.

    Алгоритм: функция последовательно извлекает последнюю цифру числа
    (через остаток от деления на 10, i % 10) и добавляет её к началу
    строки-результата, после чего число делится на 10 (i // 10),
    далее цикл повторяется. Так продолжается, пока число не станет равно 0.

    Параметры:
        - i (int): целое неотрицательное число.

    Сложность: O(N), где N — количество цифр в числе i """

    digits = "0123456789"
    if i == 0:
        return "0"
    result = ""
    while i > 0:
        result = digits[i % 10] + result
        i = i // 10
    return result
```
